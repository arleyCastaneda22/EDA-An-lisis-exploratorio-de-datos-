### Manipulación y filtrado de datos

```python
df_netflix["release_year"].value_counts()
```

```python
filtrar_shows=(df_netflix["country"]=="India") & (df_netflix["release_year"]>2010)

#Aplicar filtro y crear un nuevo dataframe
df_netflix_filtrados =df_netflix[filtrar_shows]

df_netflix_filtrados
```

```python
#Eliminar la columna "show_id"
df_netflix_filtrados=df_netflix_filtrados.drop(columns="show_id")
df_netflix_filtrados.head()
```

```python
df_netflix_filtrados=df_netflix_filtrados.fillna("N/A")
df_netflix_filtrados.head()
```

### ¿Cuántas películas y series hay en el datasets?

```python
contar_por_tipo =df_netflix_filtrados.groupby("type").size()
contar_por_tipo
```
