### Manipulación y filtrado de datos

```python
df_netflix["release_year"].value_counts()
```

```python
filtrar_shows=df_netflix["release_year"]>2010

#Aplicar filtro y crear un nuevo dataframe
df_netflix_filtrados =df_netflix[filtrar_shows]

df_netflix_filtrados
```

```python
#Eliminar la columna "show_id"
df_netflix_filtrados=df_netflix_filtrados.drop(columns="show_id")
df_netflix_filtrados.head()
```

### ¿Cuántas películas y series hay en el datasets?

```python
contar_por_tipo =df_netflix_filtrados.groupby("type").size()
contar_por_tipo
```

### ¿Cuál es el año con más estrenos en la plataforma?

```python
df_netflix_filtrados["release_year"].value_counts().sort_index(ascending=False)
```

```python
df_netflix_sin_duplicados=df_netflix_filtrados.drop_duplicates(subset="title")

df_netflix_sin_duplicados
```

### Darle formato a las fechas 

```python
df_netflix_sin_duplicados["Fecha_con_formato"]=df_netflix_sin_duplicados["date_added"].str.strip()
df_netflix_sin_duplicados["Fecha_con_formato"]=pd.to_datetime(df_netflix_sin_duplicados["Fecha_con_formato"], format='%B %d, %Y')
df_netflix_sin_duplicados
```

```python
df_netflix_sin_duplicados=df_netflix_sin_duplicados.fillna("N/A")
df_netflix_sin_duplicados
```

### Eliminar columna de fecha con formato no deseado

```python
df_con_formato_de_fecha_correcto=df_netflix_sin_duplicados.drop(columns="date_added")
df_con_formato_de_fecha_correcto
```

### Renombrar columna de fecha

```python
df_con_formato_de_fecha_correcto.rename(columns={"Fecha_con_formato":"date_added"}, inplace=True)   
df_con_formato_de_fecha_correcto
```

### Generos más populares 

```python
generos_mas_populares=df_con_formato_de_fecha_correcto.groupby("listed_in").size().sort_values(ascending=False)
generos_mas_populares
```

### ¿Cuál es el país que más contenido ha producido?

```python
pais_mas_contenido=df_con_formato_de_fecha_correcto.groupby("country").agg({"title":"count"}).sort_values(by="title",ascending=False)
pais_mas_contenido = pais_mas_contenido.rename(columns={"title":"cantidad"})
pais_mas_contenido
```

### ¿Qué directores han dirigido más contenido en la plataforma?

```python
directores_mas_contenido = df_con_formato_de_fecha_correcto.groupby("director").agg({
    "title": "count",
    "date_added": "min", 
    "rating": lambda x: x.mode().iat[0] if not x.mode().empty else None  
}).sort_values(by="title", ascending=False)

# Renombrar columnas para mayor claridad
directores_mas_contenido = directores_mas_contenido.rename(columns={"title": "Cantidad_Contenido", "date_added": "Primera_Fecha"})
directores_mas_contenido
```

### ¿Cuáles son los actores más frecuentes en el casting?

```python
actores_mas_cotizados =df_con_formato_de_fecha_correcto["cast"].str.split(",").explode()
actores_mas_cotizados =actores_mas_cotizados.groupby(actores_mas_cotizados).size().sort_values(ascending=False)
actores_mas_cotizados.head(30)
```
